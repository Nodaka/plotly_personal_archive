```python
import plotly.graph_objs as go
import plotly.io as pio
pio.kaleido.scope.mathjax = None
import kaleido
import numpy as np
```

```python
my_layout = go.Layout(
    autosize=False,
    #colorway=[],
    #font=dict(family = 'Segoe UI', size=16, color='black'),
    width=800,
    height=500,
    plot_bgcolor='rgba(0,0,0,0)',
    margin=dict(l=65, r=15, b=65, t=15, pad=0, autoexpand=False),
    xaxis=dict(
        title=dict(text='x axis', font=dict(size=24 ,family='Arial')),
        tickfont=dict(size=20, family='Arial'),
        zeroline=False,
        showline=True,
        linewidth=1,
        linecolor='black',
        mirror=True,
        #tick0=0,
        #dtick=0.1,
        #tickvals=[],
        #range=(0,1),
        gridcolor='black',
        gridwidth=1,
        griddash='dot',
        ),
    yaxis=dict(
        title=dict(text='y axis', font=dict(size=24 ,family='Arial')),
        tickfont=dict(size=20, family='Arial'),
        zeroline=False,
        showline=True,
        linewidth=1,
        linecolor='black',
        mirror=True,
        #tick0=0,
        #dtick=0.1,
        #tickvals=[],
        #range=(0,1),
        gridcolor='black',
        gridwidth=1,
        griddash='dot',
        ),
    )
my_layout_template = go.layout.Template(layout=my_layout)
pio.templates["my_template"] = my_layout_template
#pio.templates.default = 'my_template'
```

```python
np.random.seed(1)

N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5

fig = go.Figure(layout = go.Layout(template=pio.templates["my_template"]))
fig.add_trace(go.Scatter(x=random_x, y=random_y0,
                    mode='lines',
                    name='lines'))
fig.add_trace(go.Scatter(x=random_x, y=random_y1,
                    mode='lines+markers',
                    name='lines+markers'))
fig.add_trace(go.Scatter(x=random_x, y=random_y2,
                    mode='markers', name='markers'))
fig.write_image('test.png', scale=1, engine='orca')
fig.show()
```
